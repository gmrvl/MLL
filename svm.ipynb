```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
import random
```

```python
X, y = make_blobs(n_samples=100, centers=2, random_state=10)
clf = svm.SVC(kernel='linear') # 
clf.fit(X, y) 
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)
# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
plt.draw()
plt.show()

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
x_n = random.uniform(x_min, x_max)
y_n = random.uniform(y_min, y_max)

arr = np.append(X, [[x_n,y_n]], axis = 0)
res = clf.predict([[x_n,y_n]])
print(res)
plt.scatter(arr[:,0], arr[:,1], c=np.append(y, 3), s=30, cmap=plt.cm.Paired)
ax = plt.gca()
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
plt.draw()
plt.show()

plt.scatter(arr[:,0], arr[:,1], c=np.append(y, res), s=30, cmap=plt.cm.Paired)
ax = plt.gca()
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
plt.draw()
plt.show()
```
